## Plot the Color-Magnitude Diagram

To plot data with errorbars, we'll use a slightly different plotting method in matplotlib, called `plt.errorbar`. It can be used as follows:

`plt.errorbar(x_data, y_data, xerr = x_errors, yerr = y_errors, marker = 'o', linestyle='None')`

In the cell below, plot the V vs. V-R color-magnitude digram for your, adjusting the axes as necessary, and adding all relevant labels (title, axis labels, legend).  Remember lower the number in magnitude, the brighter the object! 

```python
# Use pandas to read in your cluster data file as a new dataframe. HINT: use "delim_whitespace = True".
clustername = 'NGC663'
cluster = pd.read_csv(datadir + clustername + 'cluster_photometry_final.csv')


BminusV = cluster['B-V']
BminusVerr = cluster['B-V_err']

B = cluster['B_mcal']
V = cluster['V_mcal']
R = cluster['R_mcal']

# VminusR = V-R
# BminusV = B-V

Verr = cluster['V_mcal_err']
Berr = cluster['B_mcal_err']
Rerr = cluster['R_mcal_err']

# VminusRerr = np.sqrt(Verr**2. + Rerr**2.)
# BminusVerr = np.sqrt(Berr**2. + Verr**2.)


# your CMD here:
plt.figure(figsize=(10,10))
plt.errorbar(BminusV, V, xerr=BminusVerr, yerr=Verr, marker='o', color='k',linestyle='None', alpha=0.4)
# plt.xlim(-0.5, 1.3)
plt.gca().invert_yaxis()
plt.title(clustername, fontsize=24)
plt.xlabel('color', fontsize=24)
plt.ylabel('brightness', fontsize=24)
plt.xlim(-1,2)
plt.ylim(16, 6.5)
# plt.savefig(f'{datadir}{clustername}_CMD.pdf')
plt.show()
```
